### 有些tweet的评分的分母不为10

#### 定义
- 通过pandas的正则表达式和extract手动重新提取评分

#### 代码


```python
data_1_clean.rating_denominator = data_1_clean.text.str.extract(r'\D+[\d\.]+/([1-9]+0)',expand=False)
data_1_clean.rating_numerator = data_1_clean.text.str.extract(r'\D+([\d\.]+)/[1-9]+0',expand=False)
```

```python
data_1_clean.rating_numerator.value_counts()
```

```python
data_1_clean.rating_denominator.value_counts()
```

```python
data_1_clean.rating_numerator = data_1_clean.rating_numerator.astype('float64')
data_1_clean.rating_denominator = data_1_clean.rating_denominator.astype('float64')
```

```python
data_1_clean.rating_denominator.dropna(inplace = True)
data_1_clean.rating_numerator.dropna(inplace = True)
```

```python
data_1_clean.info()
```

#### 使用lamda表达式将某些对多个狗的评分整除后计算

```python
data_1_clean['rate']= data_1_clean.rating_numerator / data_1_clean.rating_denominator
data_1_clean.rating_numerator = data_1_clean['rate'] * 10
data_1_clean.rating_denominator = 10.0
```

```python
def clear(x):
    a =x.is_integer()
    return(a)    
data_1_clean= data_1_clean[data_1_clean.rating_numerator.map(clear)]
```

```python
data_1_clean = data_1_clean[~data_1_clean.rating_numerator.isna()]
```

```python
melt_df = data_1_clean[["tweet_id","doggo","floofer","pupper","puppo"]]
melt_df = pd.melt(melt_df, id_vars=["tweet_id"],value_vars = ["doggo","floofer","pupper","puppo"], var_name="dog",value_name = 'dog_stage').drop(columns= 'dog')
```

```python
melt_df = melt_df[melt_df["dog_stage"] != "None"]
```
